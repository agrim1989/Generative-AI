# Mastering Tokenization with NLTK

This notebook explores different tokenization techniques provided by the Natural Language Toolkit (NLTK) in Python. Tokenization is the process of breaking down text into smaller units called tokens, which can be words, sentences, or even subword units. It's a fundamental step in most Natural Language Processing (NLP) tasks.

## 1. Setup: Installation and Imports

First, we need to ensure NLTK is installed and import the necessary modules. While NLTK is often pre-installed in Colab, including `pip install` is good practice. We'll import `nltk` itself, along with specific tokenizers like `sent_tokenize`, `word_tokenize`, `TreebankWordTokenizer`, and `RegexpTokenizer`.

```
# Install NLTK (uncomment if not already installed)
# !pip install nltk

import nltk
from nltk.tokenize import sent_tokenize, word_tokenize, TreebankWordTokenizer, RegexpTokenizer

print("NLTK and various tokenizers imported successfully!")
```

## 2. Define Our Corpus (Text Data)

A 'corpus' is a collection of text documents. For this demonstration, we'll use a single paragraph of text as our input corpus. We specify its type as `str` (string).

```
corpus: str = "The quick brown fox jumps over the lazy dog. This is an example sentence. What about numbers like 123.45 and special characters like @#$?! NLTK is powerful!"

print(f"Original Corpus:\n\n{corpus}\n")
```

## 3. Download NLTK Data

Many NLTK functions, especially tokenizers, rely on pre-trained models or data files. The `punkt` tokenizer data is commonly required for both sentence and some word tokenization tasks. We'll download it to ensure our tokenizers work correctly.

```
# Download the 'punkt' tokenizer models
# This is essential for sent_tokenize and some word_tokenize functionalities.
nltk.download('punkt', quiet=True)

print("'punkt' tokenizer data downloaded successfully!")
```

## 4. Sentence Tokenization (`sent_tokenize`)

Sentence tokenization is the process of splitting a text into individual sentences. NLTK's `sent_tokenize` function uses an unsupervised algorithm (trained on the 'punkt' data) to identify sentence boundaries.

```
"""Splits the corpus into a list of sentences using NLTK's sent_tokenize."""
sentences: list[str] = sent_tokenize(corpus)

print("--- Sentence Tokenization Results ---")
print(f"Number of sentences: {len(sentences)}\n")
for i, sentence in enumerate(sentences):
    print(f"Sentence {i+1}: {sentence}")
print("\n")
```

## 5. Word Tokenization (Generic `word_tokenize`)

Word tokenization splits sentences or paragraphs into individual words or punctuation marks. NLTK's `word_tokenize` is a general-purpose tokenizer, often leveraging the Treebank tokenizer internally.

```
"""Splits the corpus into a list of words using NLTK's generic word_tokenize."""
words_general: list[str] = word_tokenize(corpus)

print("--- General Word Tokenization Results ---")
print(f"Number of tokens: {len(words_general)}\n")
print(words_general)
print("\n")
```

## 6. Treebank Word Tokenization (`TreebankWordTokenizer`)

The Treebank tokenizer is widely used in NLP and follows specific rules to handle contractions, punctuation, and other linguistic phenomena. It's often considered the standard for English word tokenization.

```
"""Splits the corpus into a list of words using NLTK's TreebankWordTokenizer."""
treebank_tokenizer = TreebankWordTokenizer()
words_treebank: list[str] = treebank_tokenizer.tokenize(corpus)

print("--- Treebank Word Tokenization Results ---")
print(f"Number of tokens: {len(words_treebank)}\n")
print(words_treebank)
print("\n")
```

## 7. Regular Expression Tokenization (`RegexpTokenizer`)

`RegexpTokenizer` allows for highly flexible tokenization based on regular expressions. This is useful when you need to define custom rules for what constitutes a token, such as only alphanumeric words, or specific patterns.

```
"""Tokenizes the corpus based on a regular expression, matching only alphanumeric words."""
# This regex matches sequences of alphanumeric characters (words)
regex_tokenizer = RegexpTokenizer(r'\w+')
words_regex: list[str] = regex_tokenizer.tokenize(corpus)

print("--- Regular Expression Word Tokenization (alphanumeric words) Results ---")
print(f"Number of tokens: {len(words_regex)}\n")
print(words_regex)
print("\n")
```
